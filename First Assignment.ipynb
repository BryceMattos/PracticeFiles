### My code in the next line will show an example of the wrong way to make copies of a list in Python since it will edit the previously existing list. It then shows three proper practices (that I know) for creating list copies that won't change previous lists when edited.

```python
# Wrong way to copy
test_list = [1, 2, 3, 4, 5, 6, 7, 8, 9]

list_copy = test_list
list_copy.append("wrong way to copy a list")
print("The test list now reflects:\n", test_list, sep = "")

#Correct ways to copy
test_list = [1, 2, 3, 4, 5, 6, 7, 8, 9]

list_copy = test_list[:] #first correct way to copy a list
list_copy2 = test_list.copy() #second correct way to copy a list
list_copy3 = [i for i in test_list] #third correct way to copy a list

list_copy.append("This should not change the test list")
list_copy2.append("This should not change the test list")
list_copy3.append("This should not change the test list")

print("\nThe test list now reflects:\n", test_list, sep = "")
print("\nThe first list copy now reflects:\n", list_copy, sep = "")
print("The second list copy now reflects:\n", list_copy2, sep = "")
print("The third list copy now reflects:\n", list_copy3, sep = "")
```
